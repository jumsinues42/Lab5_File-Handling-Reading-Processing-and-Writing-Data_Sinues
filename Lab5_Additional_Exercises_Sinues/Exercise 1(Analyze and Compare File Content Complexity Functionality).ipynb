```python
from pathlib import Path
import csv

def analyze_complexity():
    # Define the student activity folder
    folder_path = Path.home() / "Documents" / "Activity_5_Files"
    
    # Automatically load all .txt files in the folder [cite: 248]
    txt_files = list(folder_path.glob("*.txt"))
    stats = []

    for file_path in txt_files:
        text = file_path.read_text(encoding="utf-8")
        lines = [line for line in text.splitlines() if line.strip()] # Ignore empty lines [cite: 249]
        words = text.split()
        chars = len("".join(words)) # Characters excluding whitespace [cite: 249]

        total_lines = len(lines) if lines else 1
        total_words = len(words)
        
        # Compute Metrics [cite: 250]
        words_per_line = total_words / total_lines
        chars_per_word = chars / total_words if total_words > 0 else 0

        stats.append({
            "Filename": file_path.name,
            "Lines": total_lines,
            "Words": total_words,
            "Characters": chars,
            "Words/Line": round(words_per_line, 2),
            "Chars/Word": round(chars_per_word, 2)
        })

    # Sort files based on word density (highest to lowest) [cite: 250]
    stats.sort(key=lambda x: x["Words/Line"], reverse=True)

    # Output tabular summary [cite: 251]
    print(f"{'Filename':<25} | {'Lines':<5} | {'Words':<5} | {'Words/Line':<10}")
    print("-" * 55)
    for s in stats:
        print(f"{s['Filename']:<25} | {s['Lines']:<5} | {s['Words']:<5} | {s['Words/Line']:<10}")

    # Export to CSV [cite: 253]
    csv_path = folder_path / "content_summary.csv"
    with open(csv_path, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=stats[0].keys())
        writer.writeheader()
        writer.writerows(stats)

analyze_complexity()
```
