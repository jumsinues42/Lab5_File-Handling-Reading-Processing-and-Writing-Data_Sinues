```python
from pathlib import Path
from datetime import datetime
import shutil

def smart_backup_system(target_filename):
    # Student-specific variables
    student_id = "TUPM-25-4168"
    student_name = "Jumar Sinues"
    
    # 1. Define paths (Matches your VS Code sidebar)
    base_dir = Path.home() / "Documents" / "Activity_5_Files"
    target_file = base_dir / target_filename
    
    # Check if the source file exists
    if not target_file.exists():
        print(f"Error: {target_filename} not found in {base_dir}")
        return

    # 2. Create the timestamped backup file
    backup_dir = base_dir / f"backup_{student_id}"
    backup_dir.mkdir(parents=True, exist_ok=True)
    
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    backup_name = f"{target_file.stem}_{student_id}_{timestamp}{target_file.suffix}"
    backup_path = backup_dir / backup_name
    
    # Perform the copy
    shutil.copy2(target_file, backup_path)
    file_size = target_file.stat().st_size
    
    # 3. Construct the Log Entry for the text file
    log_entry = (
        f"--- Backup Entry ---\n"
        f"Student: {student_name} ({student_id})\n"
        f"Original: {target_filename}\n"
        f"Timestamp: {timestamp}\n"
        f"Size: {file_size} bytes\n"
        f"Path: {backup_path}\n"
        f"{'-'*40}\n"
    )
    
    # Update/Create the continuously growing log file
    log_file = base_dir / f"backup_log_{student_id}.txt"
    with open(log_file, "a", encoding="utf-8") as log:
        log.write(log_entry)
        
    # 4. DISPLAYED CONFIRMATION (Exact format from module)
    print(log_entry) # Prints the metadata block
    print(f"Backup completed for {student_id} ({student_name})")
    print(f"File saved as: {backup_name}")
    print(f"Log updated at: Documents/Activity_5_Files/backup_{student_id}/")

# Execution
smart_backup_system("intro_TUPM-25-4168.txt")
```
