```python
from pathlib import Path
import collections
import re

def word_frequency_analyzer():
    folder_path = Path.home() / "Documents" / "Activity_5_Files"
    student_id = "2025-XXXX" # Replace with your ID
    
    # Define common stopwords to exclude [cite: 256]
    stopwords = {"the", "is", "and", "to", "in", "a", "of", "it", "with", "for"}
    
    overall_counts = collections.Counter()
    txt_files = list(folder_path.glob("*.txt"))

    for file_path in txt_files:
        # Read and normalize (lowercase, remove punctuation) [cite: 257]
        content = file_path.read_text(encoding="utf-8").lower()
        clean_text = re.sub(r'[^\w\s]', '', content)
        words = clean_text.split()
        
        # Filter stopwords [cite: 258]
        meaningful_words = [w for w in words if w not in stopwords]
        overall_counts.update(meaningful_words)

    # Output most frequent meaningful words [cite: 260]
    print("\nTop 10 Meaningful Words:")
    for word, count in overall_counts.most_common(10):
        print(f"{word}: {count}")

word_frequency_analyzer()
```
